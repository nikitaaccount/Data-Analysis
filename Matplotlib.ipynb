```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Pie Chart

```python
areas=['Marketing','Sales','Development','HR','Customer Support']
budget=[2.5,4,10,1,6]
```

```python
plt.pie(budget,labels=areas,autopct="%0.1f%%") # autopct for percentage
plt.show()
```

### Scatter Plot

```python
x=[24,45,12,49,97,32,11,44]
y=[45,34,67,87,55,22,89,90]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary Graph')
```

 ### 2D Plot

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[233,456,770,120,222,444,200,300]
offline_price=[400,300,100,333,444,565,565,899]
chor_bazaar=[100,200,300,400,150,250,450,500]

plt.plot(product_id,online_price,color="green",marker='D',markersize=5,label="Online Prices")
plt.plot(product_id,offline_price,color="red",markersize=10,label="Offline Prices")
plt.plot(product_id,chor_bazaar,color="purple",marker='+',markersize=10,label="Chori wale Prices")

plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.legend() 
plt.show()# information will be removed
```

### Bar Graph

```python
year=[2012,2013,2014,2015,2016,2017,2018]
price=[65000,69000,75000,55000,52000,64000,81000]

plt.bar(year,price,color="red")
plt.xlabel("Year")
plt.ylabel("Price")
plt.title("Year By Year Comp")
plt.show()
```
