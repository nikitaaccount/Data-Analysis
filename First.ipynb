### Sort values (applicable  to both series and dataframes)

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False)
```

```python
data.sort_values('city') # for making changes permanent write inplace =True
```

### Drop duplicates method

```python
data.drop_duplicates(subset=['city'])
```

```python
data.drop_duplicates(subset=['city','season']).shape
```

```python
# to find the last matches of every season
data.drop_duplicates('season',keep='last')
```

```python
# to find the winner of every season
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```
